```python
from transformers import pipeline
```

```python
model_name = "pysentimiento/bertweet-pt-sentiment"
analyzer = pipeline("sentiment-analysis", model=model_name)

exemplos = [
    "Eu amei muito este dia, foi maravilhoso!",           # 1-Frase simples positiva
    "Este serviço é uma porcaria, nunca mais volto.",     # 2-Frase simples negativa
    "O banco estava muito longe.",                        # 3-Ambiguidade (Banco  objeto ou instituição)
    "Eu não desgosto nem um pouco dessa ideia.",          # 4-Dupla negativa
    "Nossa, que atendimento rápido demorou só 2 horas!",  # 5-Ironia/Sarcasmo
    "O filme é tecnicamente bom mas muito cansativo.",    # 6-Contraste de opinião
    "Mano que bagulho doido que aconteceu hoje ",         # 7-Gíria
    "O céu está nublado hoje em São Paulo.",              # 8-Frase neutra/informativa
    "Estou morrendo de rir com esse vídeo.",              # 9-Expressão idiomática
    "A comida estava ok nada de especial."                #10-Emoção neutra
]


print(f"{'Texto':<55} | {'Label':<10} | Score")
for texto in exemplos:
    resultado = analyzer(texto)[0]
    label = resultado['label']
    score = resultado['score']
    print(f"{texto:<55} | {label:<10} | {score:}")
```
